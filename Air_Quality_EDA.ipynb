```python
# Exploratory Data Analysis (EDA) for All Files

## Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
city_day = pd.read_csv('city_day.csv')
city_hour = pd.read_csv('city_hour.csv')
station_day = pd.read_csv('station_day.csv')
station_hour = pd.read_csv('station_hour.csv', low_memory=False)
stations = pd.read_csv('stations.csv')
```

```python
## EDA for city_day dataset
print("City Day Dataset Info:")
print(city_day.info())
```

```python
print(city_day.describe())
```

```python
sns.heatmap(city_day.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values in city_day")
plt.show()
```

```python
sns.histplot(city_day['AQI'].dropna(), bins=30, kde=True)
plt.title("AQI Distribution in city_day")
plt.show()
```

```python
city_day['Date'] = pd.to_datetime(city_day['Date'])
city_day.groupby('Date')['AQI'].mean().plot(figsize=(14,6))
plt.title("City-level Average AQI Over Time")
plt.show()
```

```python
## EDA for city_hour dataset
print("City Hour Dataset Info:")
print(city_hour.info())
```

```python
print(city_hour.describe())
```

```python
sns.heatmap(city_hour.isnull(), cbar=False, cmap='plasma')
plt.title("Missing Values in city_hour")
plt.show()
```

```python
sns.histplot(city_hour['AQI'].dropna(), bins=30, kde=True)
plt.title("AQI Distribution in city_hour")
plt.show()
```

```python
## EDA for stations dataset (metadata)
print("Stations Dataset Info:")
print(stations.info())
```

```python
print(stations.describe())
```

```python
# Count of stations by city
top_cities = stations['City'].value_counts().head(10)
top_cities.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Cities with Most Stations")
plt.xlabel("City")
plt.ylabel("Number of Stations")
plt.show()
```

```python
stations['City'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Cities with Most Air Quality Stations')
plt.xlabel('City')
plt.ylabel('Number of Stations')
plt.show()
```

```python
stations['State'].value_counts().head(10).plot(kind='bar', figsize=(10,5), color='orange')
plt.title('Top 10 States with Most Air Quality Stations')
plt.xlabel('State')
plt.ylabel('Number of Stations')
plt.show()
```
